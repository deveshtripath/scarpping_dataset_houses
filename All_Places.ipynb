```python
final=pd.DataFrame()
page = 1
plot = []
city= []
money = []
unit = []
rup_per_area = []
unitarea = []
type = []
side_open = []
desc=[]
apartement = []
project_Name = []

while page != 136:
    
    webpage=requests.get("https://www.makaan.com/listings?cityId=25&listingType=buy&propertyType=apartment&page={page}").text
    soup=BeautifulSoup(webpage,'lxml')


    try:
        formats_on_page = soup.select("div.title-line a.typelink")
        for product_format in formats_on_page:
            plot.append(product_format.get_text())
    except:
        plot.append(np.nan)

    try:
        formats_on_page = soup.select("span.cityName")
        for product_format in formats_on_page:
            city.append(product_format.get_text())
    except:
        city.append(np.nan)
   
    try:
        formats_on_page = soup.select("td.price")
        for product_format in formats_on_page:
            money.append(product_format.get_text())
    except:
        money.append(np.nan)
        
    try:
        formats_on_page = soup.select("tr.hcol td.val")
        for product_format in formats_on_page:
            type.append(product_format.get_text()) 
    except:
        type.append(np.nan)

    try:
        formats_on_page = soup.select("span.unit")
        for product_format in formats_on_page:
            unit.append(product_format.get_text()) 
    except:
        unit.append(np.nan)
        
    try:
        formats_on_page = soup.select("tr.hcol td.rate")
        for product_format in formats_on_page:
            rup_per_area.append(product_format.get_text())
    except:
        rup_per_area.append(np.nan)
        
    try:
        formats_on_page = soup.select("tr.hcol td.size")
        for product_format in formats_on_page:
            unitarea.append(product_format.get_text())
    except:
        unitarea.append(np.nan)
    
    try:
        company=soup.find_all('ul',class_='listing-details')
        for i in company:
            side_open.append(i.find_all('li',class_='keypoint')[0].text.strip())
    except:
        side_open.append(np.nan)
    
    try:
        formats_on_page = soup.select("div.listing-description")
        for product_format in formats_on_page:
            desc.append(product_format.get_text())
    except:
        desc.append(np.nan)
    
    
    try:
        formats_on_page = soup.select("a.typelink")
        for product_format in formats_on_page:
            apartement.append(product_format.get_text())
    except:
        apartement.append(np.nan)
    
    
    try:
        formats_on_page = soup.select("a.projName")
        for product_format in formats_on_page:
            project_Name.append(product_format.get_text())
    except:
        project_Name.append(np.nan)
    
    page = page + 1
    print(page)

    

df=pd.DataFrame({'Type_of_plot':plot,
       'City':city,
       'Total Price':money,
       'Unit':unit,
       'Rupee_per_area':rup_per_area,
       'Type':type,
       'Open_Date':side_open,
       'Description':desc,
        'Area' :unitarea,
        "Apartement":apartement
})

  
final=final.append(df,ignore_index=True)
```
